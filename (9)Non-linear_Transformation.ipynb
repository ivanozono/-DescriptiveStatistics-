# Non-linear Transformation

In this notebook, we will explore the concept of non-linear transformations, which are used when we have strongly skewed, non-symmetric data. These transformations are applied before linear scaling and are generally logarithmic, sigmoid, or polynomial functions.

The aim is to apply a function to non-symmetrically distributed data to make them symmetric, and then apply linear scaling. It's important to visualize the data to understand its distribution before deciding the next step. Some functions that can be applied include square root, tangent hyperbolic, and logistic functions.

These transformations are closely related to Analytic Geometry and Scaling. One of the activation functions in neural networks is the hyperbolic tangent (tgh(x)), a fact worth remembering!

Let's start by importing the necessary libraries and loading our data.

```
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load the data
# For the purpose of this notebook, we will generate a skewed dataset using numpy
np.random.seed(0)
data = np.random.exponential(scale=2.0, size=1000)

# Convert the data to a pandas DataFrame
df = pd.DataFrame(data, columns=['Data'])

# Display the first few rows of the DataFrame
df.head()
```

We have successfully loaded our data. Now, let's visualize the distribution of our data using a histogram.

```
# Plotting a histogram of the data
plt.figure(figsize=(10, 6))
sns.histplot(df['Data'], kde=True, bins=30)
plt.title('Histogram of Data')
plt.xlabel('Data')
plt.ylabel('Frequency')
plt.show()
```

From the histogram, we can see that our data is right-skewed. This is a common situation where we would apply a non-linear transformation to make the data more symmetric for further analysis.

Let's apply a log transformation to our data and visualize the result.

```
# Apply a log transformation to the data
df['Log_Data'] = np.log(df['Data'])

# Plotting a histogram of the transformed data
plt.figure(figsize=(10, 6))
sns.histplot(df['Log_Data'], kde=True, bins=30)
plt.title('Histogram of Log Transformed Data')
plt.xlabel('Log Transformed Data')
plt.ylabel('Frequency')
plt.show()
```

After applying the log transformation, we can see from the histogram that the data is now more symmetric and less skewed. This is a common technique used in data preprocessing to handle skewed data.

In this notebook, we have learned about non-linear transformations and their applications in data preprocessing. We have also seen how to apply a log transformation to skewed data to make it more symmetric for further analysis.
