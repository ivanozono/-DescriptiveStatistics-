# Matriz de Covarianza

In this notebook, we will explore the concept of the covariance matrix, a fundamental mathematical tool in the field of data science and artificial intelligence. The covariance matrix is a measure that indicates the amount and type of dependence between multiple variables.

## Key Concepts

1. **Definition of the covariance matrix**: The covariance matrix is a square matrix that contains the covariances between pairs of variables. The entry in the i-th row and the j-th column of the matrix is the covariance between the i-th and the j-th variable.

2. **Importance of the covariance matrix**: The covariance matrix is an essential tool in statistics and data science. It is used in Principal Component Analysis (PCA), a dimensionality reduction technique. It is also useful for understanding the relationships between different variables in a dataset.

3. **Calculation of the covariance matrix in Python**: We will use numpy, matplotlib, seaborn and sklearn libraries to load the 'iris' dataset, standardize its features and calculate the covariance matrix.

4. **Visualization of the covariance matrix**: We will use seaborn to visualize the covariance matrix as a heatmap. This provides a visual representation of the relationships between the variables.

Let's get started!

```
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Load the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

The above code block imports the necessary libraries and loads the iris dataset. The iris dataset is a multivariate dataset introduced by the British statistician and biologist Ronald Fisher in his 1936 paper. This dataset became a typical test case for many statistical classification techniques in machine learning.

The dataset contains 50 samples from each of three species of Iris flowers (Iris setosa, Iris virginica, and Iris versicolor). Four features were measured from each sample: the lengths and the widths of the sepals and petals.

Next, let's visualize the relationships between different pairs of these features using a pairplot.

```
# Visualizing the relationships between different pairs of features
sns.pairplot(iris, hue = 'species')
```

The pairplot above provides a visual representation of the relationships between different pairs of features in the iris dataset. Each plot in the diagonal represents the distribution of a single feature (sepal length, sepal width, petal length, and petal width) for three species of Iris flowers. The other plots represent the scatter plots between pairs of these features.

Next, let's standardize the features of the dataset. Standardization of a dataset is a common requirement for many machine learning estimators: they might behave badly if the individual features do not more or less look like standard normally distributed data (e.g., Gaussian with 0 mean and unit variance).

```
# Standardizing the features
scaler = StandardScaler()
scaled = scaler.fit_transform(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
)
scaled
```

## Step 2: Load the Data

In this step, we will load the Titanic dataset from the seaborn library. The dataset contains passenger information from the Titanic, including whether or not they survived, their age, fare, class, sex, and more.

```
import seaborn as sns

# Load the dataset
ndata = sns.load_dataset('titanic')

# Display the first 5 rows of the dataframe
ndata.head()
```

## Step 3: Data Exploration

In this step, we will explore the Titanic dataset to understand the data better. We will check the data types of the columns, look for missing values, and generate descriptive statistics of the dataset.

```
# Check the data types of the columns
ndata.info()

# Check for missing values
ndata.isnull().sum()

# Generate descriptive statistics
ndata.describe(include='all')
```

## Step 4: Data Cleaning

In this step, we will clean the Titanic dataset. We will fill the missing values in the 'age' column with the median age, and we will drop the 'deck' column because it has too many missing values. We will also convert the 'sex' column to a binary column where 0 represents 'male' and 1 represents 'female'.

```
# Fill the missing values in the 'age' column with the median age
ndata['age'].fillna(ndata['age'].median(), inplace=True)

# Drop the 'deck' column
ndata.drop('deck', axis=1, inplace=True)

# Convert the 'sex' column to a binary column
ndata['sex'] = ndata['sex'].map({'male': 0, 'female': 1})

# Check the first 5 rows of the cleaned dataframe
ndata.head()
```

## Step 5: Data Analysis

In this step, we will analyze the Titanic dataset. We will generate a correlation matrix to understand the relationships between different features. We will also visualize the survival rate by sex and class.

```
# Generate a correlation matrix
ndata.corr()

# Visualize the survival rate by sex and class
sns.catplot(x='class', y='survived', hue='sex', data=ndata, kind='bar')
```

## Step 6: Conclusion

In this notebook, we explored the Titanic dataset. We cleaned the data, filled missing values, converted categorical data to numerical data, and analyzed the data. We found that the survival rate is higher for females and for passengers in higher classes. This analysis could be further expanded by exploring other features and creating predictive models.
