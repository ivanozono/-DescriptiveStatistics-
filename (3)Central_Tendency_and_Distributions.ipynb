# Estadística descriptiva y visualización de distribuciones con Python

En este cuaderno, exploraremos las medidas de tendencia central y cómo visualizar distribuciones univariadas y bivariadas utilizando la biblioteca Seaborn en Python.

```
# Import necessary libraries
import numpy as np
import seaborn as sns

# Load the penguins dataset
penguins = sns.load_dataset("penguins")

# Display the first few rows of the dataset
penguins.head()
```

## Medidas de tendencia central

Las medidas de tendencia central son valores que representan el centro de un conjunto de datos. Incluyen la media, la mediana y la moda.

```
# Calculate the mean, median, and mode of the 'flipper_length_mm' column
mean = penguins['flipper_length_mm'].mean()
median = penguins['flipper_length_mm'].median()
mode = penguins['flipper_length_mm'].mode()[0]

mean, median, mode
```

## Visualización de distribuciones univariadas

Seaborn proporciona varias formas de visualizar distribuciones univariadas, incluyendo histogramas, gráficos de densidad del kernel y gráficos de función de distribución acumulativa empírica.

```
# Visualize a histogram of the 'flipper_length_mm' column
sns.displot(penguins, x='flipper_length_mm')
```

```
# Visualize a kernel density estimate (KDE) of the 'flipper_length_mm' column
sns.displot(penguins, x='flipper_length_mm', kind='kde')
```

```
# Visualize an empirical cumulative distribution function (ECDF) of the 'flipper_length_mm' column
sns.displot(penguins, x='flipper_length_mm', kind='ecdf')
```

## Visualización de distribuciones bivariadas

Seaborn también permite visualizar la relación entre dos variables. Vamos a visualizar la relación entre 'bill_length_mm' y 'bill_depth_mm'.

```
# Visualize a bivariate distribution of 'bill_length_mm' and 'bill_depth_mm'
sns.displot(penguins, x='bill_length_mm', y='bill_depth_mm')
```
